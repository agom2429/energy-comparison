```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url ='https://en.wikipedia.org/wiki/World_energy_supply_and_consumption'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
#creating a dataframe with the websites table but excluding the foot of the table 
table_index = 0  
table = soup.find_all('table')[table_index]
df_from_website = pd.read_html(str(table))[0]
df_from_website = df_from_website.iloc[:-1]
new_df = pd.DataFrame(columns=df_from_website.columns)
for index, row in df_from_website.iterrows():
    new_df.loc[index] = row
    
new_df
```
